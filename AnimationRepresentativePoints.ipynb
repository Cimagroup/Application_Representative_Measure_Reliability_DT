```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Generar puntos aleatorios
np.random.seed(42)
points = np.random.rand(1000, 2)  # Nube de 1000 puntos en 2D

# Porcentaje final de puntos (20%)
final_percentage = 0.2
final_count = int(len(points) * final_percentage)

# Seleccionar de antemano los puntos que se mantendrán al final
final_indices = np.random.choice(len(points), final_count, replace=False)

# Crear figura para la animación
fig, ax = plt.subplots()
sc = ax.scatter(points[:, 0], points[:, 1], alpha=0.6)

# Ajustar límites del gráfico
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title("Selecting representative points")

# Lista de índices de puntos a eliminar progresivamente
remaining_indices = list(set(range(len(points))) - set(final_indices))
np.random.shuffle(remaining_indices)

# Número total de frames
total_frames = 50

# Función de actualización para la animación
def update(frame):
    # Calcular cuántos puntos deben mantenerse en este frame
    step = len(remaining_indices) // total_frames
    if step == 0:
        step = 1  # Asegurar que siempre eliminemos algo si el conjunto es pequeño
    
    indices_to_keep = remaining_indices[frame * step:] + list(final_indices)

    # Actualizar los puntos mostrados
    sc.set_offsets(points[indices_to_keep])
    return sc,

# Crear la animación
ani = animation.FuncAnimation(fig, update, frames=total_frames, interval=100)

plt.show()

```

```python
ani.save("puntos_representativos.gif")
```
